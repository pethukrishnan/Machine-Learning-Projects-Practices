## One hot Encoding

```python
data_oh = data

def one_hot(data,column):
    data = pd.concat([data,pd.get_dummies(data[column],prefix=column,drop_first=True,dtype='int64')],axis=1)
    data = data.drop([column],axis=1)
    return data

cols = ['season','month','hour','holiday','weekday','workingday','weather']

for col in cols:
    data_oh = one_hot(data_oh,col)
data_oh.head()
```

## Input Split

```python
X = data_oh.drop(columns = ['atemp','windspeed','casual','registered','count'],axis=1)
Y = data_oh['count']
```

## Model Training

```python
from sklearn.linear_model import LinearRegression,Ridge,HuberRegressor,ElasticNetCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import ExtraTreesRegressor,RandomForestRegressor,GradientBoostingRegressor

models = [LinearRegression(),
          Ridge(),
          HuberRegressor(),
          ElasticNetCV(),
          DecisionTreeRegressor(),
          ExtraTreesRegressor(),
          RandomForestRegressor(),
          GradientBoostingRegressor()]
```

```python
from sklearn import model_selection
def train(model):
    k_fold = model_selection.KFold(n_splits=5)
    pred = model_selection.cross_val_score(model,X,Y,cv=k_fold,scoring='neg_mean_squared_error')
    cv_score = pred.mean()
    print('Model: ',model)
    print('CV Score: ',abs(cv_score))
    
```

```python
for model in models:
    train(model)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
# plot the error difference

error = y_test - y_pred
fig,ax = plt.subplots()
ax.scatter(y_test,error)
ax.axhline(lw=3,color='black')
ax.set_xlabel('Observed')
ax.set_ylabel('Error')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test,y_pred))
```
